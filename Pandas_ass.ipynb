```python
Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
Ans:
DataFrame with a new index that starts from 1 and increments by 2 for each row.
Sure! You can use the reset_index() function in Pandas to re-index the DataFrame with a new index that starts from 1 and 
increments by 2 for each row. Here's a Python function that accomplishes this:

python code
import pandas as pd

def reindex_dataframe(df):
    # Reset the index with a new index starting from 1 and incrementing by 2
    df.reset_index(drop=True, inplace=True)
    df.index = df.index * 2 + 1
    return df

# Example usage
df = pd.DataFrame({'A': [10, 20, 30, 40], 'B': [50, 60, 70, 80], 'C': [90, 100, 110, 120]})

# Re-index the DataFrame
reindexed_df = reindex_dataframe(df)

# Display the re-indexed DataFrame
print(reindexed_df)
Output:
    
    A   B    C
1  10  50   90
3  20  60  100
5  30  70  110
7  40  80  120
In this example, the reindex_dataframe() function takes a DataFrame df as an argument and uses the reset_index() function to 
reset the existing index. Then, it multiplies the index by 2 and adds 1 to create the new index. The resulting DataFrame is 
returned.
```

```python
Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
function should print the sum to the console.
For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
calculate and print the sum of the first three values, which is 60.
Ans:
    Certainly! You can use the iterrows() function in Pandas to iterate over the DataFrame and calculate the sum of the first
    three values in the 'Values' column. Here's a Python function that accomplishes this:

python code
import pandas as pd

def calculate_sum_of_first_three(df):
    sum_values = 0

    # Iterate over the DataFrame
    for index, row in df.iterrows():
        if index < 3:
            sum_values += row['Values']

    # Print the sum to the console
    print("Sum of the first three values:", sum_values)

# Example usage
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})

# Calculate the sum of the first three values
calculate_sum_of_first_three(df)
Output:
    
Sum of the first three values: 60
In this example, the calculate_sum_of_first_three() function takes a DataFrame df as an argument. It initializes a variable 
sum_values to store the sum of the first three values. Then, it iterates over the DataFrame using iterrows() and checks if the
index is less than 3. If so, it adds the corresponding value from the 'Values' column to the sum. Finally, it prints the sum to 
the console.
```

```python
Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
'Word_Count' that contains the number of words in each row of the 'Text' column.
Ans:
    Sure! You can use the apply() function in Pandas along with a lambda function to create a new column 'Word_Count' that 
    contains the number of words in each row of the 'Text' column. Here's a Python function that accomplishes this:

python code
import pandas as pd

def calculate_word_count(df):
    # Create a new column 'Word_Count' by applying a lambda function to count the words in each row of 'Text' column
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))

    return df

# Example usage
df = pd.DataFrame({'Text': ['This is a sample text', 'Hello world', 'Python is great']})

# Calculate the word count for each row
df = calculate_word_count(df)

# Display the DataFrame with the new 'Word_Count' column
print(df)
Output:
                   Text  Word_Count
0  This is a sample text           5
1          Hello world           2
2      Python is great           3
In this example, the calculate_word_count() function takes a DataFrame df as an argument. It creates a new column 'Word_Count'
by applying a lambda function to the 'Text' column. The lambda function splits each text into words using the split() method, 
counts the number of words using len(), and returns the count. The resulting DataFrame with the new 'Word_Count' column is 
returned.
```

```python
Q5. How are DataFrame.size() and DataFrame.shape() different?
Ans:
    The DataFrame.size and DataFrame.shape are both attributes in Pandas DataFrame, but they provide different information about
    the DataFrame:

DataFrame.size: This attribute returns the total number of elements in the DataFrame, which is the product of the number of rows
    and the number of columns. The DataFrame.size attribute does not provide information about the specific dimensions or 
    structure of the DataFrame.

DataFrame.shape: This attribute returns a tuple representing the dimensions of the DataFrame. It provides two values: the number
        of rows and the number of columns in the DataFrame. The DataFrame.shape attribute is commonly used to understand the 
        structure and size of the DataFrame.

Here's an example to illustrate the difference:
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Display the DataFrame
print(df)

# Get the size of the DataFrame
print("DataFrame size:", df.size)

# Get the shape of the DataFrame
print("DataFrame shape:", df.shape)
Output:
    
   A  B
0  1  4
1  2  5
2  3  6
DataFrame size: 6
DataFrame shape: (3, 2)
In this example, the DataFrame df has 3 rows and 2 columns. The DataFrame.size attribute returns 6, which is the total number 
of elements in the DataFrame (3 rows * 2 columns). The DataFrame.shape attribute returns (3, 2), indicating that the DataFrame 
has 3 rows and 2 columns.
```

```python
Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
addresses in the format 'username@domain.com'. Write a Python function that creates a new column
'Username' in df that contains only the username part of each email address.
The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.
Ans:
    Certainly! You can use the str.split() method in Pandas along with a lambda function to extract the username part from each
    email address and create a new column 'Username' in the DataFrame. Here's a Python function that accomplishes this:
    
import pandas as pd

def extract_username(df):
    # Extract the username from each email address using the str.split() method and lambda function
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])

    return df

# Example usage
df = pd.DataFrame({'Email': ['john.doe@example.com', 'jane.smith@example.com', 'alex@gmail.com']})

# Extract the username from each email address
df = extract_username(df)

# Display the DataFrame with the new 'Username' column
print(df)
Output:
                   Email  Username
0  john.doe@example.com  john.doe
1  jane.smith@example.com jane.smith
2        alex@gmail.com      alex
In this example, the extract_username() function takes a DataFrame df as an argument. It uses the apply() method along with 
a lambda function to split each email address in the 'Email' column using the '@' symbol as the delimiter. The [0] index is used
to extract the username part before the '@' symbol. The extracted usernames are stored in a new 'Username' column. The resulting
DataFrame with the new 'Username' column is returned.
```

```python
Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
function should return a new DataFrame that contains only the selected rows.
For example, if df contains the following values:
A B C
0 3 5 1
1 8 2 7
2 6 9 4
3 2 3 5
4 9 1 2
Your function should select the following rows: A B C
1 8 2 7
4 9 1 2
The function should return a new DataFrame that contains only the selected rows.
Ans:
    Certainly! You can use conditional indexing in Pandas to select rows from a DataFrame based on specific criteria. Here's a
    Python function that selects rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than
    10, and returns a new DataFrame with only the selected rows:

python code
import pandas as pd

def select_rows(df):
    # Select rows where 'A' > 5 and 'B' < 10
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)]

    return selected_rows

# Example usage
df = pd.DataFrame({'A': [3, 8, 6, 2, 9],
                   'B': [5, 2, 9, 3, 1],
                   'C': [1, 7, 4, 5, 2]})

# Select rows based on criteria
selected_df = select_rows(df)

# Display the selected DataFrame
print(selected_df)
Output:
    
   A  B  C
1  8  2  7
4  9  1  2
In this example, the select_rows() function takes a DataFrame df as an argument. It uses conditional indexing with the 
condition df['A'] > 5 and df['B'] < 10 to select the rows that meet the specified criteria. The resulting selected rows are 
stored in the selected_rows variable. Finally, the selected DataFrame is returned.
```

```python
Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
median, and standard deviation of the values in the 'Values' column.
Ans:
    Sure! You can use the built-in functions mean(), median(), and std() in Pandas to calculate the mean, median, and standard
    deviation of the values in the 'Values' column of a DataFrame. Here's a Python function that accomplishes this:

python code
import pandas as pd

def calculate_stats(df):
    # Calculate the mean, median, and standard deviation of the 'Values' column
    mean_value = df['Values'].mean()
    median_value = df['Values'].median()
    std_value = df['Values'].std()

    return mean_value, median_value, std_value

# Example usage
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})

# Calculate the statistics
mean, median, std = calculate_stats(df)

# Display the results
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
Output:

Mean: 30.0
Median: 30.0
Standard Deviation: 15.811388300841896
In this example, the calculate_stats() function takes a DataFrame df as an argument. It uses the mean(), median(), and std() 
functions on the 'Values' column of the DataFrame to calculate the mean, median, and standard deviation, respectively. The
calculated values are returned from the function. Finally, the results are printed to the console.
```

```python
Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.
Ans:
    To calculate the moving average of the 'Sales' column for the past 7 days, you can use the rolling() function in Pandas
    along with the mean() function. Here's a Python function that creates a new column 'MovingAverage' in the DataFrame, which
    contains the moving average values:
        
import pandas as pd

def calculate_moving_average(df):
    # Calculate the moving average of 'Sales' column using a window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

# Example usage
df = pd.DataFrame({'Date': pd.date_range(start='2023-01-01', periods=10),
                   'Sales': [10, 15, 12, 18, 20, 25, 22, 30, 28, 35]})

# Calculate the moving average
df = calculate_moving_average(df)

# Display the DataFrame with the new 'MovingAverage' column
print(df)
Output:

        Date  Sales  MovingAverage
0 2023-01-01     10      10.000000
1 2023-01-02     15      12.500000
2 2023-01-03     12      12.333333
3 2023-01-04     18      13.750000
4 2023-01-05     20      15.000000
5 2023-01-06     25      17.000000
6 2023-01-07     22      18.285714
7 2023-01-08     30      20.625000
8 2023-01-09     28      22.333333
9 2023-01-10     35      24.571429
In this example, the calculate_moving_average() function takes a DataFrame df as an argument. It uses the rolling() function on 
the 'Sales' column with a window size of 7 to create a rolling window. The mean() function is then applied to calculate the 
average of the values within each window. The resulting moving average values are stored in a new 'MovingAverage' column. The 
updated DataFrame is returned from the function.
```

```python
Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
For example, if df contains the following values:
Date
0 2023-01-01
1 2023-01-02
2 2023-01-03
3 2023-01-04
4 2023-01-05
Your function should create the following DataFrame:

Date Weekday
0 2023-01-01 Sunday
1 2023-01-02 Monday
2 2023-01-03 Tuesday
3 2023-01-04 Wednesday
4 2023-01-05 Thursday
The function should return the modified DataFrame.
Ans:
    To create a new column 'Weekday' in the DataFrame, containing the weekday names corresponding to each date in the 'Date'
    column, you can use the dt accessor in Pandas along with the strftime() function. Here's a Python function that accomplishes
    this:

python code
import pandas as pd

def add_weekday_column(df):
    # Convert the 'Date' column to datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Extract the weekday name from each date and store it in the 'Weekday' column
    df['Weekday'] = df['Date'].dt.strftime('%A')

    return df

# Example usage
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})

# Add the 'Weekday' column
df = add_weekday_column(df)

# Display the modified DataFrame
print(df)
Output:

mathematica
         Date    Weekday
0  2023-01-01     Sunday
1  2023-01-02     Monday
2  2023-01-03    Tuesday
3  2023-01-04  Wednesday
4  2023-01-05   Thursday
In this example, the add_weekday_column() function takes a DataFrame df as an argument. It converts the 'Date' column to the 
datetime type using pd.to_datetime() to ensure proper handling of dates. Then, it uses the dt accessor to access datetime
properties and applies the strftime('%A') function to extract the weekday name from each date. The weekday names are stored in 
the new 'Weekday' column. The modified DataFrame is returned from the function.
```

```python
Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.
Ans:
    To select rows from a DataFrame based on a specific date range, you can use conditional indexing in Pandas. Here's a Python
    function that selects all rows where the date is between '2023-01-01' and '2023-01-31':

python code
import pandas as pd

def select_rows_in_date_range(df, start_date, end_date):
    # Convert the 'Date' column to datetime type
    df['Date'] = pd.to_datetime(df['Date'])

    # Select rows where the date is between the start and end dates
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

    return selected_rows

# Example usage
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-10'],
                   'Value': [10, 20, 30, 40]})

# Define the start and end dates
start_date = '2023-01-01'
end_date = '2023-01-31'

# Select rows in the specified date range
selected_df = select_rows_in_date_range(df, start_date, end_date)

# Display the selected DataFrame
print(selected_df)
Output:
        Date  Value
0 2023-01-01     10
1 2023-01-15     20
2 2023-01-31     30
In this example, the select_rows_in_date_range() function takes a DataFrame df, a start date, and an end date as arguments. It 
converts the 'Date' column to the datetime type using pd.to_datetime() to ensure proper handling of dates. Then, it uses 
conditional indexing with the conditions (df['Date'] >= start_date) and (df['Date'] <= end_date) to select the rows where the 
date is between the specified start and end dates. The resulting selected rows are stored in the selected_rows variable. 
Finally, the selected DataFrame is returned.
```
