```python
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
Ans:
    In pandas, the loc and iloc functions are used to access and manipulate data in a DataFrame, but they differ in their 
    indexing methods. Here's a breakdown of their differences:

loc (label-based indexing):

The loc function allows indexing and slicing based on the labels of rows and columns.
It uses explicit labels or boolean arrays for indexing.
The syntax for using loc is df.loc[row_indexer, column_indexer].
The row and column indexers can be single labels, lists of labels, or boolean arrays.
When using loc, both the start and end indices are inclusive.
The labels used in loc can be any valid label present in the index or column names.
iloc (integer-based indexing):

The iloc function allows indexing and slicing based on the integer positions of rows and columns.
It uses implicit integer positions for indexing.
The syntax for using iloc is df.iloc[row_indexer, column_indexer].
The row and column indexers can be single integers, lists of integers, or boolean arrays.
When using iloc, the start index is inclusive, but the end index is exclusive.
The positions used in iloc are integer positions, starting from 0 for the first row/column.
In summary, loc uses label-based indexing, allowing you to access data using explicit labels or boolean arrays. On the other 
hand, iloc uses integer-based indexing, allowing you to access data using implicit integer positions.
```

```python
Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
Ans:
    In pandas, window functions allow you to perform calculations on a specific "window" or subset of data within a DataFrame.
    These functions operate on a rolling or expanding window of data and can be used for tasks such as calculating moving
    averages, cumulative sums, or other window-based calculations.

Window functions in pandas can be categorized into two types:

Rolling Window Functions:

Rolling window functions operate on a fixed-size window that moves along the data.
They calculate aggregate values over that window at each point.
The window can be defined based on a fixed number of rows or a time period.
Some commonly used rolling window functions include:
rolling(): Creates a rolling window object.
mean(): Calculates the mean of the values within the window.
sum(): Calculates the sum of the values within the window.
std(): Calculates the standard deviation of the values within the window.
min(): Calculates the minimum value within the window.
max(): Calculates the maximum value within the window.
Expanding Window Functions:

Expanding window functions include all the data from the start to the current point.
They calculate aggregate values for the expanding window over time.
The size of the window increases as more data is included.
Some commonly used expanding window functions include:
expanding(): Creates an expanding window object.
mean(): Calculates the mean of all the values up to the current point.
sum(): Calculates the sum of all the values up to the current point.
std(): Calculates the standard deviation of all the values up to the current point.
min(): Calculates the minimum value from the start to the current point.
max(): Calculates the maximum value from the start to the current point.
These window functions are powerful tools for analyzing time series data or performing calculations on sliding windows of data 
in pandas. They provide flexibility in defining the window size, allowing you to capture and analyze specific subsets of your 
data based on your requirements.
```

```python
Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.
Ans:
    Certainly! Here's a Python program that reads a CSV file containing sales data for different products and visualizes the 
    data using a stacked bar chart to show the sales of each product category over time:
    
import pandas as pd
import matplotlib.pyplot as plt

# Prompt the user to enter the file path
file_path = input("Enter the CSV file path: ")

# Read the CSV file
df = pd.read_csv(file_path)

# Convert the 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Group the data by 'Date' and 'Product Category' and sum the sales
grouped_data = df.groupby(['Date', 'Product Category'])['Sales'].sum().unstack()

# Plot the stacked bar chart
grouped_data.plot(kind='bar', stacked=True)

# Set the labels and title
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales of Each Product Category Over Time')

# Display the chart
plt.show()
In this program, the user is prompted to enter the file path of the CSV file. The pd.read_csv() function is used to read the CSV
file into a DataFrame.

The 'Date' column is converted to datetime using pd.to_datetime() to ensure proper handling of date values.

The data is then grouped by 'Date' and 'Product Category' using groupby() and the sales for each category are summed using sum()
. The resulting grouped data is reshaped into a format suitable for a stacked bar chart using unstack().

The stacked bar chart is created using grouped_data.plot(kind='bar', stacked=True).

Labels and title are set using plt.xlabel(), plt.ylabel(), and plt.title().

Finally, the chart is displayed using plt.show().

Make sure the CSV file is in the correct format and accessible from the provided file path for the program to read it 
successfully. Also, ensure that the necessary libraries, such as pandas and matplotlib, are installed.
```
